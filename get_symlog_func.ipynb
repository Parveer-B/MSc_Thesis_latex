```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(-200, 200, 100000)
symlog = np.sign(a) * np.log1p(np.abs(a))

fig, (ax1, ax2) = plt.subplots(1, 2, dpi = 300, figsize = (20, 5))
ax1.plot(a, symlog)
ax1.set_ylim(-20, 20)
ax1.set_xlim(-20, 20)
ax2.plot(a[50001:], symlog[50001:])
ax2.set_xscale('log')
plt.show()
```
